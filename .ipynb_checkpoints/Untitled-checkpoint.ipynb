```python
import geopandas as gpd
import numpy as np
import pandas as pd
import shapely as sp
import matplotlib.pyplot as plt
from shapely.geometry import Point, LineString, Polygon
import random
import networkx as nx
import osmnx as ox
import seaborn as sns
import requests, time
import folium
import datetime
```

```python
ox.settings.useful_tags_way = ["highway", "oneway", "length"]
```

```python
def calc_dist(G, orig_coords, dest):
    dest_x = G.nodes[dest]["x"]
    dest_y = G.nodes[dest]["y"]
    return round(ox.distance.great_circle(orig_coords[1], orig_coords[0], dest_y, dest_x))

def get_rand_walk(G, orig_coords, node, max_length, length=0, visited=None, tol=-20):
    if(length >= max_length):
        return [node]

    if visited is None:
        visited = set()
    visited.add(node)

    neighbors = list(set(G.adj[node]) - visited)
    if(len(neighbors) == 0):
        return [node]

    filtered_neighbors = []
    for neighbor in neighbors:
        if(calc_dist(G, orig_coords, neighbor) - calc_dist(G, orig_coords, node) > tol):
            filtered_neighbors.append(neighbor)

    if(len(filtered_neighbors) == 0):
        return [node]

    while(len(filtered_neighbors) > 0):
        rand_idx = np.random.randint(0,len(filtered_neighbors))
        neighbor = filtered_neighbors[rand_idx]
        subpath = get_rand_walk(G, orig_coords, neighbor, max_length, 
                                 length + sum(ox.routing.route_to_gdf(G, [node, neighbor])["length"]), visited, tol)
        if(len(subpath) == 0):
            filtered_neighbors.pop(rand_idx)
            
        if len(subpath) >= 1:
            subpath_length = sum(ox.routing.route_to_gdf(G, [node] + subpath)["length"])
            if subpath_length + length >= max_length:
                return [node] + subpath
            else:
                filtered_neighbors.pop(rand_idx)

    return []
```

```python
def calc_dist_nodes(orig, dest):
    orig_x = G.nodes[orig]["x"]
    orig_y = G.nodes[orig]["y"]
    dest_x = G.nodes[dest]["x"]
    dest_y = G.nodes[dest]["y"]
    return round(ox.distance.great_circle(orig_y, orig_x, dest_y, dest_x))

def get_shortest_path(node, dest, path_len=0, visited=None):
    if(node == dest):
        return True, path_len, [node]

    if visited is None:
        visited = set()
        path_len = 0
    visited.add(node)

    neighbors = list(set(G.adj[node]) - visited)
    if(len(neighbors) == 0):
        return False, path_len, [node]

    path_lengths = []
    good_neighbors = []
    good_paths = []
    while(len(neighbors) > 0):
        neighbor = random.choice(neighbors)
        next_path_len = path_len + calc_dist_nodes(node, neighbor)
        reached, path_len, path = get_shortest_path(neighbor, dest, next_path_len, visited)
        print(reached,path_len,path)
        if(reached):
            path_lengths.append(path_len)
            good_neighbors.append(neighbor)
            good_paths.append(path)
            continue
        else:
            neighbors = list(set(neighbors) - set([neighbor]))
            continue

    if(len(path_lengths) > 0):
        assert(len(good_neighbors)==len(path_lengths) and len(good_paths) == len(path_lengths))
        min_len = 10000000
        min_len_node = None
        min_path = None
        for i in range(len(path_lengths)):
            if(path_lengths[i]<min_len):
                min_len = path_lengths[i]
                min_len_node = good_neighbors[i]
                min_path = good_paths[i]
                
        return True, min_len, [min_len_node] + min_path    
```

```python
def get_route(orig_point_obj, num_paths, walk_radius):
    path = []
    route_nodes = []
    path_len = 0 # meters
    G = None
    
    for i in range(num_paths):
        if i>0 and len(path)>0:
            start_node = G.nodes[path[-1]]
            start_point_obj = Point(start_node['x'], start_node['y'])
        else:
            start_point_obj = orig_point_obj
    
        start_time = datetime.datetime.now()
        lon,lat = start_point_obj.coords[0]
        G = ox.graph_from_point((lat,lon), dist=walk_radius, network_type="walk")
        print(f"OSMX (subpath {i}) {(datetime.datetime.now() - start_time).total_seconds()} s")
    
        start_lon, start_lat = start_point_obj.coords[0]
        start_pt = ox.distance.nearest_nodes(G, X=start_lon, Y=start_lat)
    
        subpath = get_rand_walk(G, orig_point_obj.coords[0], start_pt, walk_radius/2)
        if(len(subpath) < 2):
            print(f"subpath {i}: 0 mi")
            path_len = path_len + 0
        else:
            print(f"subpath {i}: {sum(ox.routing.route_to_gdf(G, subpath)["length"])/1609} mi")
            path = path + subpath
            path_len = path_len + sum(ox.routing.route_to_gdf(G, subpath)["length"])
            route_nodes = route_nodes + [G.nodes[n] for n in subpath]
            
    return path, path_len, route_nodes
```

```python
gdf = gpd.read_file("data/ne_10m_admin_0_countries.shp")
gdf_airports = gpd.read_file("data/ne_10m_airports.shp")
gdf=gdf.set_index('ADMIN')
gdf = gdf.set_geometry("geometry")
```

```python
hotel_radius = 5*1609
walk_radius = 16*1609
num_paths = 6
min_path_len = (num_paths-1) * walk_radius
```

```python
journey = dict()
num_trials = 2
for trial in range(num_trials):
    rand_admin_idx = random.choice(admin_indices)
    filtered_airports = sp.intersection(gdf_airports.geometry,gdf.iloc[rand_admin_idx].geometry)
    country_airports_pts = [point for point in filtered_airports if not sp.is_empty(point)]
    
    path = []
    path_len = 0
    while(len(country_airports_pts) > 0):
        airport = random.choice(country_airports_pts)
        gdf_point = gpd.GeoDataFrame(geometry=[airport], crs="EPSG:4326")
        gdf_projected = gdf_point.to_crs("EPSG:3857")
        polygon = gdf_projected.buffer(hotel_radius).to_crs("EPSG:4326").union_all()
        gdf_hotels = ox.features_from_polygon(polygon, tags={"tourism": "hotel"})
        if(len(list(gdf_hotels.geometry)) == 0):
            country_airports_pts.remove(airport)
    
        hotel = gdf_hotels.iloc[np.random.randint(0,gdf_hotels.shape[0])].geometry.centroid
        path, path_len, route_nodes = get_route(hotel, num_paths, walk_radius)
        if(path_len < min_path_len):
            country_airports_pts.remove(airport)
            continue
        else:
            journey['route'] = LineString([(n['x'], n['y']) for n in route_nodes])
            journey['hotel'] = hotel
            journey['airport'] = airport
            break
            
    if(len(country_airports_pts) == 0):
        admin_indices.remove(rand_admin_idx)
        continue
    else:
        break
```

```python
checkpoints_gdf = gpd.GeoDataFrame(
    {"name": ["airport", "hotel"]},
    geometry=[journey['airport'], journey['hotel']],
    crs="EPSG:4326"
)
route_gdf = gpd.GeoDataFrame(
    {"name": ["route"]},
    geometry=[journey['route']],
    crs="EPSG:4326"
)

m = gdf.explore(legend=False, tooltip=False, popup=False, color="lightgrey",
                highlight_kwds={"fillOpacity": 0, "weight": 1},
                style_kwds={"fillOpacity": 0.3, "weight": 0.5},
                name="World"
)

checkpoints_gdf.explore(m=m, color="green", name="checkpoints")

lon, lat = journey['route'].centroid.coords[0]
route_gdf.explore(m=m, color="red", 
                  style_kwds={"weight": 4, "opacity": 0.8},
                  map_kwds={"location": [lon,lat], "zoom_start":2},
                  name="route"
)
```
