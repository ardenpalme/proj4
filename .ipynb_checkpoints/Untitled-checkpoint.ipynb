```python
def calc_dist(orig, dest):
    orig_x = G.nodes[orig]["x"]
    orig_y = G.nodes[orig]["y"]
    dest_x = G.nodes[dest]["x"]
    dest_y = G.nodes[dest]["y"]
    return round(ox.distance.great_circle(orig_y, orig_x, dest_y, dest_x))

def get_rand_path(origin, node, max_length, length=0, visited=None):
    if(length >= max_length):
        return [node]

    if visited is None:
        visited = set()
    visited.add(node)

    neighbors = list(set(G.successors(node)) - visited)
    if(len(neighbors) == 0):
        return [node]

    filtered_neighbors = []
    for neighbor in neighbors:
        if(calc_dist(origin, neighbor) > calc_dist(origin, node)):
            filtered_neighbors.append(neighbor)

    if(len(filtered_neighbors) == 0):
        return [node]
    
    for neighbor in filtered_neighbors:
        subpath = get_rand_path(origin, neighbor, max_length, 
                                 length + sum(ox.routing.route_to_gdf(G, [node, neighbor])["length"]), visited)
        if(len(subpath) == 0):
            continue
            
        if len(subpath) >= 1:
            subpath_length = sum(ox.routing.route_to_gdf(G, [node] + subpath)["length"])
            if subpath_length + length >= max_length:
                return [node] + subpath
            else:
                continue

    return []
```
